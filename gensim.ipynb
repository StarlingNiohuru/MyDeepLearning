```python
from gensim.models import Word2Vec
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
import re
import numpy as np
```

```python
with open('D://人类简史.txt', 'r', encoding='utf-8') as f:
          raw_text = f.read()
text = raw_text.replace('\r', '').replace('\n', '').replace('\t', '').lower()
partern = '[\u4e00-\u9fa5]|[，、。！？；：“”《》——【】『』]|[-.%/]|[A-Za-z]+'
sequences = re.findall(partern, text)
```

```python
ss = np.asanyarray(sequences).reshape(-1,20)
print(ss.shape)
```

```python
sss=[]
for s in ss:
    sss.append(list(s))
print(sss[4:6])
```

```python
model = Word2Vec(sss, window=20,min_count=1,iter=100)
model.wv.save_word2vec_format('D:\word2vec_human.txt')
```

```python
model.similar_by_word('国')
```
